```python
test=pd.read_csv("test.csv")
test
```

```python
xtest=train.iloc[:,0]
ytest=train.iloc[:,1]
```

```python
plt.scatter(xtest,ytest)
plt.title("feature vs label")
plt.show()
```

```python
theta0=0
theta1=0
alpha=0.0001
epochs=10000
```

```python
m=float(len(x))
m
```

```python
for i in range(epochs):
    h_x=theta0+theta1*x
    print(h_x.iloc[0] - y.iloc[0])
    d_theta1=(1/m)*sum(x*(h_x -y))
    d_theta0=(1/m)*sum(h_x -y)
    theta1=theta1 - alpha *d_theta1
    theta1=theta0 - alpha *d_theta0
```

```python
y_pred=theta0+theta1*xtest
plt.scatter(xtest,ytest)
plt.plot([min(xtest),max(xtest)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```
